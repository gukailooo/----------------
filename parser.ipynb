```python
import requests
from bs4 import BeautifulSoup

# URL страницы с таблицей
url = 'https://en.wikipedia.org/wiki/List_of_mathematical_constants#cite_ref-11'

# Выполнение HTTP-запроса для получения HTML-кода страницы
response = requests.get(url)
response.raise_for_status()  # Проверка успешности запроса

# Парсинг HTML-кода с помощью BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

tables = soup.find_all('table')
print(f"Найдено таблиц: {len(tables)}")

for i, tbl in enumerate(tables):
    print(f"Таблица {i}: классы = {tbl.get('class')}")

table_index = 0  # Замените на правильный индекс после анализа
table = tables[table_index]
```

```python
def fibonacci_recursive(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci_recursive(n-1) + fibonacci_recursive(n-2)

array = []
n = 10
for i in range(n+1):
    array.append(fibonacci_recursive(i))
print(array)
for i in array:
    print(i, end='')
```
